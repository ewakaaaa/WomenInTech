## 1. Konfiguracja z pliku `.env`

Model i klucz ustawiasz **w pliku `.env`** (`LLM_BASE_URL`, `LLM_API_KEY`,
`LLM_MODEL`). Ten notebook tylko go **wczytuje** — klucz nie jest wpisywany
w komórce, więc nie trafi przypadkiem do repo.

```python
import os

# Wczytaj konfigurację LLM z pliku .env (nic nie wpisujemy na sztywno).
from dotenv import load_dotenv

load_dotenv()

print("model:    ", os.environ.get("LLM_MODEL"))
print("base_url: ", os.environ.get("LLM_BASE_URL"))
print("klucz z .env wczytany:", bool(os.environ.get("LLM_API_KEY")))
```

## 2. Test — czy działa?

Proste wywołanie ze **structured output** (Pydantic). Jeśli wróci obiekt z polami
`country` i `capital` — konfiguracja działa.

```python
from pydantic import BaseModel
from src.llm import call_llm

class Capital(BaseModel):
    country: str
    capital: str

result = call_llm(
    messages=[{"role": "user", "content": "Jaka jest stolica Polski?"}],
    response_model=Capital,
)
print(result)
```

Jeśli powyżej zobaczyłeś coś w stylu `country='Polska' capital='Warszawa'` — działa.
